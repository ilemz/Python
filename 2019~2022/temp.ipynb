```python
import csv
import matplotlib.pyplot as plt
from matplotlib import rc

f = open('seoul.csv', 'r', encoding = 'cp949')
lines = csv.reader(f)

rc('font', family = 'NanumBarunGothic')
font1 = {'size':18, 'color':'Black'}

next(lines)

max = []
min = []
av = []

for line in lines:
  if line[4] != '':
    max.append(float(line[4]))

  if line[3] != '':
    min.append(float(line[3]))

  if line[2] != '':
    av.append(float(line[2]))


#print(max)
plt.figure(figsize = (12, 5))
plt.title('1904~2021년 기온데이터 분석', fontdict = font1)
plt.plot(max, 'r')
plt.plot(min, 'b')
plt.plot(av, 'g')
plt.show()
f.close()
```

```python
import csv
import matplotlib.pyplot as plt
from matplotlib import rc

f = open('2021_age_1.csv', 'r', encoding = 'CP949')
lines = csv.reader(f)
name = input('동이름 : ')
result = []

rc('font', family = 'NanumBarunGothic')
font1 = {'size':18, 'color':'Black'}

for line in lines:
  if name in line[0]:
    for i in line[3:]:
      result.append(int(i))

print(result)
plt.title(name+'의 인구 구조 분석', fontdict = font1)
plt.style.use('ggplot')
plt.plot(result)
plt.show()
f.close()
```

```python
import csv
import matplotlib.pyplot as plt
from matplotlib import rc

f = open('2021_gender_1.csv', 'r', encoding = 'cp949')
lines = csv.reader(f)
male = []
female = []
name = input('동이름 : ')

rc('font', family = 'NanumBarunGothic')
font1 = {'size':18, 'color':'Black'}

for line in lines:
  if name in line[0]:
    for i in line[3:104]:
      if i != '':
        male.append(int(i))

    for j in line[106:]:
      if j != '':
        female.append(int(j))

plt.title(name + '의 인구구조 분석', fontdict = font1)
plt.plot(male, label = '남성')
plt.plot(female, label = '여성')
plt.legend()
plt.show()
f.close()
```
