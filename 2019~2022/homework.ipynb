```python
import csv
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')

f = open('검색.csv', 'r', encoding = 'cp949')
lines = csv.reader(f)

next(lines)

year = ''
search = 0
Search = {}

for line in lines:
  year = line[0]
  search = int(line[1])

  Search[year] = search

print(Search)

s_list = Search.items()
s_list = sorted(s_list)
x, y = zip(*s_list)

plt.savefig('search.png')
plt.plot(x, y, 'k-')
plt.title('분기별 반려식물 검색 량')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')

f = open('종류.csv', 'r', encoding = 'cp949')
lines = csv.reader(f)
kind = ''
per = 0
Kind = {}

next(lines)

for line in lines:
  kind = line[0]
  per = float(line[1])

  Kind[kind] = per

print(Kind)

k_list = Kind.items()
k_list = sorted(k_list)
x, y = zip(*k_list)
colors = ['blue', 'skyblue', 'royalblue', 'darkblue', 'powderblue', 'turquoise']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(y, labels=x, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.title('많이 기르는 식물 종류')
plt.savefig('kind.png')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')

f = open('이유.csv', 'r', encoding = 'cp949')
lines = csv.reader(f)
reason = ''
per = 0
Reason = {}

next(lines)

for line in lines:
  reason = line[0]
  per = float(line[1])

  Reason[reason] = per

print(Reason)

r_list = Reason.items()
r_list = sorted(r_list)
x, y = zip(*r_list)
colors = ['crimson', 'coral', 'red', 'darkred', 'pink', 'peach']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(y, labels=x, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.title('식물을 기르는 이유')
plt.savefig('reason.png')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')

f = open('효과.csv', 'r', encoding = 'cp949')
lines = csv.reader(f)
effect = ''
score = 0
Effect = {}

next(lines)

for line in lines:
  effect = line[0]
  score = int(line[1])

  Effect[effect] = score

print(Effect)

e_list = Effect.items()
e_list = sorted(e_list)
x, y = zip(*e_list)

plt.bar(x, y)
plt.ylabel('점수')
plt.title('식물의 효과')
plt.savefig('effect.png')
plt.show()
```
