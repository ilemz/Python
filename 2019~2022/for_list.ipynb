```python
import matplotlib.pyplot as plt

pets = ['dog', 'cat', 'fish', 'etc', 'none']
data = [35, 28, 14, 15, 23]

plt.pie(data, explode = (0, 0.2, 0, 0, 0), labels = pets, autopct = '%d%%', shadow = True, startangle = 270)

plt.title('The portion of pets')
plt.show()
```

```python
import matplotlib.pyplot as plt

date = list(range(1, 12))

min_temp = [7.1, 7.3, 8.4, 8.6, 9.3, 7.9, 8.3, 8.1, 7.5, 8, 1]
max_temp = [17.1, 17.3, 18.4, 18.6, 19.3, 17.9, 18.3, 18.1, 17.5, 18, 11]

plt.title('Min and Max Temperature')
plt.plot(date, min_temp, label = 'Min Temperature', color = 'royalblue', linestyle = '-', marker = 'o')
plt.plot(date, max_temp, label = 'Max Temperature', color = 'crimson', linestyle = ':', marker = 'x')
plt.legend(loc = 'lower left')
plt.savefig('pet.png')
plt.show()
```
