```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

plt.rc('font', family='NanumBarunGothic')

f = open('2021_age1.csv', 'r', encoding = 'CP949')
lines = csv.reader(f)

next(lines)        #제목 다음으로 이동
lines = list(lines)
list(lines)        #리스트로 변경

mn = 1
result_name = ''
result = 0

name = input('name : ')

for line in lines:
  if name in line[0]:
    home = np.array(line[3:], dtype = int)/int(line[2])
#print(home)

for line in lines:
  away = np.array(line[3:], dtype = int)/int(line[2])
  s = np.sum((home-away)**2)

  if s < mn and name not in line[0]:         #다른 동네
    mn = s
    result_name = line[0]
    result = away

print(result_name)

addr = result_name.split(' ')

plt.title(name + '과 비슷한 인구구조는 ' + addr[-2])
plt.plot(home, label = name)
plt.plot(result, label = addr[-2])
plt.legend()
plt.show()

f.close()
```
