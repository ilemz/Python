4. 2010년부터 2020년까지 전국 대학진학률(최고, 최저 지역분석)

```python
import csv

y = list(range(2010, 2021))

for yy in y:
  f = open('jinhak.csv', 'r', encoding = 'cp949')
  lines = csv.reader(f)

  max = 0
  min = 100

  max_p = ''
  min_p = ''

  for line in lines:
    if str(yy) == line[0]:
      if max < float(line[5]):
        max = float(line[5])
        max_p = line[1] + ' ' + line[2]

      if min > float(line[5]):
        min = float(line[5])
        min_p = line[1] + ' ' + line[2]

  print('{}년 최고 진학률 : {}%, 지역 : {}' .format(yy, max, max_p))
  print('{}년 최저 진학률 : {}%, 지역 : {}' .format(yy, min, min_p))
  print('\n')

f.close()
```

데이터 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

df = pd.read_csv('jinhak.csv', encoding = 'cp949')
year = list(range(2010, 2021))

for y in year:
  condition = (df['연도'] == y)

  df_filter = df[condition]

  plt.rc('font', family = 'NanumBarunGothic')
  plt.figure(figsize = (16, 6))
  plt.title(str(y) + '년 시도별 대학 진학률')
  plt.bar(df_filter['시도'], df_filter['진학률'])
  plt.show()
```
